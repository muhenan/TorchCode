# 🟢 Easy: Embedding Layer

Implement an **embedding lookup table** from scratch.

### Signature
```python
class MyEmbedding(nn.Module):
    def __init__(self, num_embeddings: int, embedding_dim: int): ...
    def forward(self, indices: Tensor) -> Tensor: ...
```

### Rules
- `self.weight`: `nn.Parameter` of shape `(num_embeddings, embedding_dim)`
- Forward: index into weight matrix — `weight[indices]`
- Do NOT use `nn.Embedding`

````python
import torch
import torch.nn as nn
````

````python
# ✏️ YOUR IMPLEMENTATION HERE

class MyEmbedding(nn.Module):
    def __init__(self, num_embeddings, embedding_dim):
        super().__init__()
        # num_embeddings is often called vocab_size in NLP:
        # one row per token in the vocabulary.
        # nn.Parameter means this tensor is a learnable model weight:
        # it will appear in model.parameters(), be saved in state_dict(),
        # and be updated by the optimizer during training.
        # If we used plain torch.randn(...), it would just be a normal tensor.
        self.weight = nn.Parameter(torch.randn(num_embeddings, embedding_dim))

    def forward(self, indices):
        # Embedding is essentially a big learnable matrix of shape
        # (vocab_size, embedding_dim) == (num_embeddings, embedding_dim),
        # where each row is one token's vector.
        # Forward does a lookup by row index: weight[indices].
        return self.weight[indices]
````

````python
# 🧪 Debug
emb = MyEmbedding(10, 4)
idx = torch.tensor([0, 3, 7])
print('Output shape:', emb(idx).shape)
print('Matches manual:', torch.equal(emb(idx)[0], emb.weight[0]))
````
